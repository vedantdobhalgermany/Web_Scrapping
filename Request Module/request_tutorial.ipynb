# 9.Working with a public API
 - API : https://jsonplaceholder.typicode.com/
 - endpoint : posts
 - perform error handling : raise_for_status()

```python
url = "https://jsonplaceholder.typicode.com/posts"
```

## 9.1
   - Make a GET request to retriece a list of posts

```python
try : 
    response = requests.get(url)
    response.raise_for_status()
except Exception as e : 
    print(e)
else:
    status_code = response.status_code
    print(f"Status Code : {status_code}")
    if status_code == 200:
        print("\Successful GET requests!")
        posts =response.json()
        for i in range(3):
            print(f"\nPost {i+1}:")
            print(posts[i])
    else: 
        print("Unsuscessful GET request!")
```

# 9.2
 - Make a POST request to submit a post
 - includes the following parameter:
   - title
   - body
   - userId

```python
new_post = {
    "title": "Sample Post",
    "body": "This is a sample post",
    "userId":101
}
```

```python
try :
    response = requests.post(url, data=new_post)
    response.raise_for_status()

except Exception as e:
    print(e)

else:
    status_code = response.status_code
    if status_code ==201:
        print(f"Status Code : {status_code}")
        print("\n Successful POST request!")
        post = response.json()
        print("\n Post:")
        print(post)
    else: 
        print(f"Status Code: {status_code}")
        print("\n Unsuccessful Request! ")
        
```
